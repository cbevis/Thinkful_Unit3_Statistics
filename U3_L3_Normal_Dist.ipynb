```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Making a standard normally distributed variable with 1000 observations
# standard deviation of 1

mean = 0
sd = 1
n = 1000
#for repeatability
np.random.seed(25)

df = pd.DataFrame({'normal': np.random.normal(mean, sd, n),'exponential': np.random.exponential(.2, n)})

# Plotting the variables in the data frame
df.hist(sharey=True)
plt.show()
```

```python
# Making two variables
rand1 = np.random.normal(50, 300, 1000)
rand2 = np.random.poisson(1, 1000)

rand1.sort()
rand2.sort()

# making a standard normally distributed variable with 1000 observations
# a mean of 0, and standard deviation of 1 that we will use as our comparison
norm = np.random.normal(0, 1, 1000)

#Sorting values in ascending order
norm.sort()
```

```python
# plotting variable rand1 aganst norm in qqplots
plt.plot(norm, rand1, 'o')
plt.show()
```

```python
# plotting teh variable rand2 against norm in qqplots
plt.plot(norm, rand2, 'o')
plt.show()
```

```python
# Plot a histogram for rand1
plt.hist(rand1, bins=20, color='c')

# Add a vertical line at the mean
plt.axvline(rand1.mean(), color='b', linestyle='solid', linewidth=2)
# Add a vertical line at one std above mean
plt.axvline(rand1.mean() + rand1.std(), color='b', linestyle='dashed', linewidth=2)
# Add a vertical line at one std below mean
plt.axvline(rand1.mean() - rand1.std(), color='b', linestyle='dashed',linewidth=2)

plt.title('rand1')
plt.show()
```

```python
# Plot a histogram for rand2
plt.hist(rand2, bins=20, color='c')

# Add a vertical line at the mean
plt.axvline(rand2.mean(), color='b', linestyle='solid', linewidth=2)
# Add a vertical line at one std above mean
plt.axvline(rand2.mean() + rand2.std(), color='b', linestyle='dashed', linewidth=2)
# Add a vertical line at one std below mean
plt.axvline(rand2.mean() - rand2.std(), color='b', linestyle='dashed',linewidth=2)

plt.title('rand2')
plt.show()
```

# Lesson 3 - Other distributions

```python
# Generate a bernoulli distribution with p = 0.5
bernoulli = np.random.binomial(1, .5, 1000)
plt.hist(bernoulli)
plt.show()
```

```python
# Generate a binomial distribution with n=20, p=0.5
binomial = np.random.binomial(20, 0.5, 1000)
plt.hist(binomial)


# Print line at mean
mean_binomial = np.mean(binomial)
plt.axvline(mean_binomial, color='c')

plt.show()
```

```python
# Generate a gamma distribution with shape =5 and scale =1
gamma = np.random.gamma(5, 1, 1000)
plt.hist(gamma)
plt.show()
```

```python
# Plot a poisson with lambda = 3
poisson = np.random.poisson(3, 1000)
plt.hist(poisson)
plt.show()
```

# Conditional Distribution

```python
# Creating a data frame to hold the simulated ecommerce data, and 
# populating it with a normally distributed variable with mean 75
# and standard deviation 25

ecommerce = pd.DataFrame({'spending': np.random.normal(75, 25, 1000)})
#ecommerce['Spending'] = np.random.normal(75, 25, 1000)

plt.hist(ecommerce['spending'])
plt.show()
```

```python
# Adding a variable with counts of number of times of visiting the site
ecommerce['visit_count'] = np.random.randint(0, 5, 1000)

# Selecting only the cases where the visit count is greater than two 
# and plotting those

plt.hist(ecommerce[ecommerce['visit_count'] > 2]['spending'])
plt.show()
```
