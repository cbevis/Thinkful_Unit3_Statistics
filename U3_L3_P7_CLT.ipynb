```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
%matplotlib inline
```

```python
# Create function to analyze population vs sample means
def pops_to_samples(n, p1, p2, size, sample_size):
    pop1 = np.random.binomial(n, p1, size)
    pop2 = np.random.binomial(n, p2, size)
    
    sample1 = np.random.choice(pop1, sample_size, replace=True)
    sample2 = np.random.choice(pop2, sample_size, replace=True)
    
    bins=list(range(0,11))
    bin_ticks = list(map(lambda x: x + 0.5, bins))
    # Make histograms of two groups

    plt.hist(sample1, alpha=0.5, label='Sample 1', bins=bins, linewidth=1, edgecolor='black')
    plt.hist(sample2, alpha=0.5, label='Sample 2', bins=bins, linewidth=1, edgecolor='black')
    plt.xticks(bin_ticks, bins)
    plt.xlabel('Number of Successes')


    plt.legend(loc='upper right')
    plt.show()
    
    print('Mean of Sample 1 is {:.02f}.'.format(sample1.mean()))
    print('Mean of Sample 2 is {:.02f}.'.format(sample2.mean()))
    print('Standard deviation of Sample 1 is {:.02f}.'.format(sample1.std()))
    print('Standard deviation of Sample 2 is {:.02f}.'.format(sample2.std()))
    
    return list(sample1), list(sample2)
```

```python
# Calculate for n=10, p pop1 = 0.2, p pop2 = 0.5, size = 10000, sample size 100
pops_to_samples(10, 0.2, 0.5, 10000, 100)
```

```python
# Increase size of samples from 100 to 1000
# I expect the mean to be closer to the true mean and for it to
# more closely represent the population.
pops_to_samples(10, 0.2, 0.5, 10000, 1000)
```

The mean of sample 1 is closer, mean of sample 2 less close.

```python
# Decrease size of samples to 20
# I expect this one to vary a lot depending on the random seed. 
pops_to_samples(10, 0.2, 0.5, 10000, 20)
```

The means vary a lot, and the mean is less close to that of the mean.

```python
# Change value of p for pop1 to 0.3.
(sample1, sample2) = pops_to_samples(10, 0.3, 0.5, 10000, 100)
sample1 = np.array(sample1)
sample2 = np.array(sample2)
```

```python
print(ttest_ind(sample2, sample1, equal_var=False))
```

```python
# Change value of p for pop1 to 0.3.
(sample1, sample2) = pops_to_samples(10, 0.4, 0.5, 10000, 100)
sample1 = np.array(sample1)
sample2 = np.array(sample2)
```

```python
print(ttest_ind(sample2, sample1, equal_var=False))
```

The statistic got smaller and the pvalue increased.  This is because the means of the two populations are now closer together, meaning that it is harder to distinguish between the two populations and the chances that they are the same population increased.

```python
# Create function to analyze population vs sample means
def pops_to_samples_new_dist(high1, low1, high2, low2, size, sample_size):
    pop1 = np.random.uniform(low1, high1, size)
    pop2 = np.random.uniform(low2, high2, size)
    
    sample1 = np.random.choice(pop1, sample_size, replace=True)
    sample2 = np.random.choice(pop2, sample_size, replace=True)

    # Make histograms of two groups

    plt.hist(sample1, alpha=0.5, label='Sample 1', linewidth=1, edgecolor='black')
    plt.hist(sample2, alpha=0.5, label='Sample 2', linewidth=1, edgecolor='black')


    plt.legend(loc='upper right')
    plt.show()
    
    print('Mean of Sample 1 is {:.02f}.'.format(sample1.mean()))
    print('Mean of Sample 2 is {:.02f}.'.format(sample2.mean()))
    print('Standard deviation of Sample 1 is {:.02f}.'.format(sample1.std()))
    print('Standard deviation of Sample 2 is {:.02f}.'.format(sample2.std()))
    
```

```python
pops_to_samples_new_dist(10, 0, 8, 6, 10000, 100)
```

The sample mean value still accurately represents the population mean value as long as the sample size is large enough.
